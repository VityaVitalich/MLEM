```python
mean_performance = [
    0.2375094495, 0.2594323414, 0.2432822486, 0.2597759604,
    0.2526188558, 0.2489927478, 0.2437550363, 0.2546333602,
    0.2516457211, 0.2480550569, 0.2519449431, 0.2483542789,
    0.2597774245, 0.2457869634, 0.2454689984, 0.2489666137,
    0.328125, 0.1666666667, 0.2239583333, 0.28125,
    0.3541666667, 0.3588541667, 0.3541666667, 0.3557291667
]

# ANISOTROPY
anisotropy = [
    0.0461, 0.08, 0.01, 0.02,
    0.102, 0.045, 0.07, 0.03,
    0.15, 0.05, 0.05, 0.11,
    0.19, 0.13, 0.13, 0.12,
    0.15, 0.12, 0.12, 0.12,
    0.07, 0.07, 0.04, 0.03
]

# INTRINSIC
intrinsic = [
    7.0395, 26, 15, 23,
    23, 35, 18.8, 37,
    15, 19, 18, 17,
    12, 13, 16, 14,
    12, 9, 9, 14,
    1, 1, 1, 1
]
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

# Calculating Rank Correlation
corr_mean_anisotropy, anis_p_value = stats.spearmanr(mean_performance, anisotropy)
corr_mean_intrinsic, intrinsic_p_value = stats.spearmanr(mean_performance, intrinsic)

# Printing the correlations
print("Rank Correlation between Mean Performance and Anisotropy:", corr_mean_anisotropy, anis_p_value)
print("Rank Correlation between Mean Performance and Intrinsic:", corr_mean_intrinsic, intrinsic_p_value)
```

```python

# Plotting
plt.figure(figsize=(6, 3))
plt.style.use('bmh') 

# Plot for Mean Performance vs Anisotropy
plt.subplot(1, 2, 1)
plt.scatter(mean_performance, anisotropy, color='blue')
plt.title(f'corr={corr_mean_anisotropy:.2f}; p-value={anis_p_value:.2f}', fontsize=10)
plt.xlabel('Normalized performance', fontsize=8)
plt.ylabel('Anisotropy', fontsize=8)

# Plot for Mean Performance vs Intrinsic
plt.subplot(1, 2, 2)
plt.scatter(mean_performance, intrinsic, color='green')
plt.title(f'corr={corr_mean_intrinsic:.2f}; p-value={intrinsic_p_value:.2f}', fontsize=10)
plt.xlabel('Normalized performance', fontsize=8)
plt.ylabel('Intrinsic dimension', fontsize=8)

plt.tight_layout()
plt.savefig('../figs/corr.pdf')
plt.show()
```
